```python
from tensorflow.keras.layers import TextVectorization
```

```python

X=df['comment_text']
y=df[df.columns[2:]].values
```

```python
X.shape
```

```python
y.shape
```

```python
X
```

```python
y
```

```python
Max_feature=200000 #number of word in vocab
vectorizer=TextVectorization(max_tokens=Max_feature,
                             output_sequence_length=1800,
                             output_mode='int')
```

```python
vectorizer.adapt(X.values)
```

```python
vectorizer.get_vocabulary()
```

```python
vectorizer("Soumesh is Best") #numbers are represting index
```

```python
vectorizer_text=vectorizer(X.values)
```

```python
vectorizer_text
```

```python
input_tensor = tf.convert_to_tensor(vectorizer_text)
labels_tensor = tf.convert_to_tensor(y)
```

```python
dataset=tf.data.Dataset.from_tensor_slices((input_tensor, labels_tensor))
dataset=dataset.cache()
dataset=dataset.shuffle(160000)
dataset=dataset.batch(16)
dataset=dataset.prefetch(8)
```

```python
dataset.as_numpy_iterator().next()
```

```python
batch_X,batch_y=dataset.as_numpy_iterator().next()
```

```python
batch_X.shape
```

```python
len(dataset)
```

```python
train=dataset.take(int(len(dataset)*.7))
val=dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test=dataset.take(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

```python
len(train),len(val),len(test)
```

```python
train_generator=train.as_numpy_iterator().next()
```

```python
train_generator
```

## CREATE A MODEL ( LSTM )

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM,Dropout,Bidirectional,Dense,Embedding
```

```python
model=Sequential()
# we have to create a embedding layer
model.add(Embedding(Max_feature+1,32)) #vocab size,32 len vector of each word

model.add(Bidirectional(LSTM(32,activation='tanh')))

model.add(Dense(128,activation='relu'))
model.add(Dense(256,activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(6,activation='sigmoid'))
```

```python
model.compile(loss="BinaryCrossentropy",optimizer="adam")
```

```python
model.summary()
```

```python
history=model.fit(train,epochs=1,validation_data=val)
```

```python
import pickle
with open("Toxic_model", 'wb') as f:
    pickle.dump(model, f)
```

## Evalute model

```python
from tensorflow.keras.metrics import Precision,Recall,CategoricalAccuracy
```

```python
pre=Precision()
re=Recall()
acc=CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
  X_true,y_true=batch
  y_hat=model.predict(X_true)

  y_true=y_true.flatten()
  y_hat=y_hat.flatten()

  pre.update_state(y_true,y_hat)
  re.update_state(y_true,y_hat)
  acc.update_state(y_true,y_hat)
```

```python
print(pre.result().numpy(),re.result().numpy(),acc.result().numpy())
```

```python
def score_comment(comment):
    vectorized_comment = vectorizer(comment)
    results = model.predict(np.array([vectorized_comment]))

    text = ''
    for idx, col in enumerate(df.columns[2:]):
        text += '{}: {}\n'.format(col, results[0][idx]>0.5)

    return text
```
